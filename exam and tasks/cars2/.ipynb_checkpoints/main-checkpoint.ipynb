```python
from datetime import *
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
df_cars = pd.read_csv('cars.csv')
df_services = pd.read_csv('services.csv')
df_receipt1 = pd.read_csv('receipt_1.csv')
df_receipt2 = pd.read_csv('receipt_2.csv')
```

```python
df_conc_receipts = pd.concat([df_receipt1, df_receipt2], ignore_index=True)
```

```python
df_merged = df_conc_receipts.merge(df_cars, how='right')
df_merged = df_merged.merge(df_services, how='right', left_on='op_num', right_on='op_num')
```

```python
df_merged
```

```python
# де невідома кількість деталей рахується вартість без них
df_merged['general_price'] = df_merged.apply(lambda row: 
                                             row['op_price'] if (row['detail_amount'] != row['detail_amount'])
                                             else row['detail_amount']*row['detail_price'] + row['op_price'], axis=1)
```

```python
print("Summary peyment for all oerations: ", df_merged['general_price'].sum())
```

```python
df_grouped_by_operations = df_merged.groupby('op_name').sum()
df_grouped_by_operations
```

```python

df_grouped_by_operations['general_price'].plot(kind='bar')
```

```python
df_merged
```

```python
def paid_sum(op):
    df_new = df_merged.groupby(['op_name', 'brand']).sum()
    df_new = df_new.loc[op, 'amount_of_operations']
    display(df_new)
    df_new.plot(kind = 'bar', xlabel=f'amount of operations for: {op}')
```

```python
widgets.interact(paid_sum, op=['change', 'clean', 'inspection', 'repairing'])
```
